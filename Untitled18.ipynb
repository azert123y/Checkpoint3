```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic-passengers.csv",delimiter =";")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['Cabin'], inplace=True, axis=1, errors='ignore')
df['Age'].fillna(df['Age'].mode(),inplace=True)
df.dropna(
    axis=0,
    how='any',
    thresh=2,
    subset=None,
    inplace=True
)
df.head()
df.tail()
```

```python
import matplotlib.pyplot as plt
df[["Age", "Sex"]].plot.box()
```

```python
from sklearn.preprocessing import LabelEncoder
sex_le = LabelEncoder()
df["Sex"] = sex_le.fit_transform(df["Sex"])
# sex_le.inverse_transform(a)
df.head()
```

```python
plt.xlabel("Ticket")
plt.ylabel("Pclass")
plt.title("correlation between Pclass and Ticket")
vc=df['Pclass'].value_counts()
vc.plot.bar(rot=30)
```

```python
import seaborn as sns
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map( df )
```

```python
#Description :
From the plot correlation map, I see that depending on the correlation between two features the color of the case changes.
So a big correlation like between the "SibSp" and "Parch" is represented with a pink/Red color (correlation of 0.38).
A small correlation is represented with another color which is blue, for example, between the "Pclass" and the "Fare", (correlation of -0.55).
This correlation is defined by a scale that goes from dark blue to red so more the correlation is bigger, more the case takes a color that approaches red.
```

```python
def f(nom):
    Title_Dictionary = {

                    "Capt":       "Officer",

                    "Col":        "Officer",

                    "Major":      "Officer",

                      "Dr":         "Officer",

                    "Rev":        "Officer",

                    "Jonkheer":   "Royalty",

                    "Don":        "Royalty",

                    "Sir" :       "Royalty",

                   "Lady" :      "Royalty",

                  "the Countess": "Royalty",

                    "Dona":       "Royalty",

                    "Mme":        "Miss",

                    "Mlle":       "Miss",

                    "Miss" :      "Miss",

                    "Ms":         "Mrs",

                    "Mr" :        "Mrs",

                    "Mrs" :       "Mrs",

                    "Master" :    "Master"

                    }
    for k,v in Title_Dictionary.items():
        if k in nom :
            return v
    return ""
df["Title"]=df["Name"].apply(f)
Title_le = LabelEncoder()
df["Title"] = Title_le.fit_transform(df["Title"])
```
